# Projeto: Análise Exploratória de Vendas (EDA)
**Objetivo:** Realizar a limpeza e análise inicial de um dataset de vendas para identificar tendências e faturamento por categoria.
**Ferramentas:** Python, Pandas, Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Configuração para exibir gráficos no notebook
%matplotlib inline
```

```python
# Criação de um dataset de exemplo (Simulando dados de uma loja)

data = {

'ID_Pedido': [1, 2, 3, 4, 5, 6],
    'Data': ['2026-01-01', '2026-01-02', '2026-01-02', '2026-01-03', '2026-01-04', '2026-01-05'],
    'Produto': ['Notebook', 'Mouse', 'Monitor', 'Notebook', 'Teclado', 'Mouse'],
    'Quantidade': [1, 5, 2, 1, 3, 10],
    'Preco_Unitario': [4500, 150, 1200, 4500, 300, 150]

}

df = pd.DataFrame(data)
df['Data'] = pd.to_datetime(df['Data']) # Convertendo para data real
df.head()
```

```python
# Calculando o faturamento total por linha
df['Faturamento_Total'] = df['Quantidade'] * df['Preco_Unitario']

print("Dataset atualizado com faturamento:")
display(df)
```

```python
# Qual produto mais faturou?
faturamento_por_produto = df.groupby('Produto')['Faturamento_Total'].sum().sort_values(ascending=False)

print("Faturamento por Produto:")
print(faturamento_por_produto)
```

```python
# Gerando um gráfico de barras rápido
faturamento_por_produto.plot(kind='bar', color='skyblue', title='Faturamento Total por Produto')
plt.ylabel('Valor em R$')
plt.xlabel('Produtos')
plt.show()
```
