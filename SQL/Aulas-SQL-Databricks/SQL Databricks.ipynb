```sql
SELECT distinct categoria
FROM vendas_produtos
```

```sql
SELECT distinct cidade, idade
FROM vendas_clientes
```

```sql
SELECT DISTINCT cidade
FROM vendas_clientes
```

```sql
SELECT * FROM vendas_produtos
WHERE cast(replace(replace(replace(preco_unitario, 'R$ ', ''), '.', ''), ',', '.') AS FLOAT) > 500
```

```sql
SELECT count(*) AS qtd_de_clientes
FROM vendas_clientes
WHERE idade !> 30



```

```sql
SELECT count(*) AS qte_de_vendas
FROM vendas_vendas
```

```sql
SELECT COUNT(DISTINCT categoria) as total_de_categorias
FROM vendas_produtos
```

```sql
SELECT COUNT(*)
FROM vendas_clientes 
WHERE IDADE BETWEEN 20 AND 30
```

```sql
SELECT mean(idade), mode(idade), avg(idade)
FROM vendas_clientes
```

```sql
SELECT * FROM vendas_clientes
WHERE primeiro_nome LIKE"A%"
```

```sql
SELECT categoria, AVG(preco_unitario) as preco_medio_unitario
FROM vendas_produtos
GROUP BY categoria
HAVING preco_medio_unitario > 500
```

```sql
SELECT
p.nome_produto,
v.id_venda,
v.data_venda



FROM vendas_vendas AS v
JOIN vendas_produtos AS p ON (v.id_produto = p.id_produto)
```

```sql
SELECT
  p.id_produto,
  p.nome_produto,
  ROUND(sum(v.quantidade * cast(replace(replace(replace(p.preco_unitario, 'R$ ', ''), '.', ''), ',', '.') AS FLOAT)), 2) AS total_vendido_por_produto
FROM vendas_vendas AS v
JOIN vendas_produtos AS p ON (v.id_produto = p.id_produto)
GROUP BY p.id_produto, p.nome_produto
ORDER BY total_vendido_por_produto DESC
LIMIT 10
```

```sql
SELECT

p.id_produto,
p.nome_produto

FROM vendas_produto AS p
LEFT JOIN vendas_vendas AS v ON (v.id_produto = p.id_produto)
WHERE v.id_produto IS NULL;
```
